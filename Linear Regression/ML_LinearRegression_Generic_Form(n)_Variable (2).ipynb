def step_GGD(data,targets,learning_rate,m):
    dataShape = data.shape
    dataTotalPointsInColumn = dataShape[0]
    dataTotalPointsInRow = dataShape[1]   
    new_m = []
    for k in range(dataTotalPointsInRow):
        newTemp_M_Sum =0
        for i in range(dataTotalPointsInColumn):
            y=targets[i][0] 
            M_Sum=0

                M_Sum += x*m[j]

```python
def g_Descent(data,targets,iterations,learning_rate):
    m=np.zeros(5)
    for i in range(iterations):
        m=step_Descent(data,targets,learning_rate,m)
        print(i,cost(data,targets,m))
        print()
    return m    
```

```python
def step_Descent(data,targets,learning_rate,m):
    X,Y = data.shape
    for i in range(Y):
        Total_Sum = 0
        for j in range(X):
            y=targets[j]
            row_Sum =0
            for k in range(Y):
                row_Sum += m[k]*data[j][k]
            Total_Sum += (-2/X)*data[j][i]*(y-row_Sum)
        m[i] -= learning_rate*Total_Sum
    return m    
```

```python
def cost(data,targets,m):
    total_cost = 0
    X,Y=data.shape
    for i in range(X):
        temp_cost = 0
        for j in range(Y):
            temp_cost += data[i][j]*m[j]  
        total_cost += ((targets[i][0] - temp_cost)**2)
    return (total_cost/X)    
```
