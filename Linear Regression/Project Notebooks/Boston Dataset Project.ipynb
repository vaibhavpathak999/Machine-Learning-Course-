#### Fetching data from csv files using  "np.genfromtxt"  method

```python
data_Train = np.genfromtxt(r"E:\MachineLearning Coding Ninjas Course\Linear Regression\Datasets\0000000000002417_training_boston_x_y_train (1).csv", delimiter=",")
data_Test = np.genfromtxt(r"E:\MachineLearning Coding Ninjas Course\Linear Regression\Datasets\0000000000002417_test_boston_x_test (1).csv", delimiter=",")
```

#### creating dataset and target set and importing model_selection for splitting data into train and test files

```python
data = data_Train[:,0:13]
target = data_Train[:,13:14]
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(data,target)
```

#### printing the shapes of split

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

### importing the Linear Regression model from sklearn library and predicting over X_test dataset

```python
from sklearn.linear_model import LinearRegression
algo = LinearRegression()
algo.fit(X_train,Y_train)
predict = algo.predict(X_test)
predict.shape
```

### Finding different types of errors using metrics "Mean Absolute Error", "Mean Squared Error", "Root Mean Squared Error"

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, predict))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, predict))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, predict)))
```

### printing data_Train and data_Train shape

### Computing the score

```python
algo.score(X_test,Y_test)
```

```python
def g_Descent(data,targets,iterations,learning_rate):
    m=np.zeros(14)
    for i in range(iterations):
        m=step_Descent(data,targets,learning_rate,m)
        print(i,cost(data,targets,m))
        print()
    return m    

#simple step gradient algorithm

def step_Descent(data,targets,learning_rate,m):
    X,Y = data.shape
    for i in range(Y):
        Total_Sum = 0
        for j in range(X):
            y=targets[j]
            row_Sum =0
            for k in range(Y):
                row_Sum += m[k]*data[j][k]
            Total_Sum += (-2/X)*data[j][i]*(y-row_Sum)
            m[i] -= learning_rate*Total_Sum # Stocashtic gradient descent
        #m[i] -= learning_rate*Total_Sum
    return m    

def cost(data,targets,m):
    total_cost = 0
    X,Y=data.shape
    for i in range(X):
        temp_cost = 0
        for j in range(Y):
            temp_cost += data[i][j]*m[j]  
        total_cost += ((targets[i][0] - temp_cost)**2)
    return (total_cost/X)    

def predict(testData,m):
    M,N =testData.shape
    predictions = []
    for i in range(M):
        y=0
        for j in range(N):
            y += m[j]*testData[i][j] 
        y += m[4]    
        predictions.append(float(y))
    return np.array(predictions)         
```
